```python
import pandas as pd
import numpy as np

from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
```

Data will be loaded and a stritified train test split will be performed

```python
df = pd.read_csv("../data/creditcard.csv")
```

```python
X = df.drop("Class", axis=1)
y = df["Class"]
```

```python
X_train, X_test, y_train, y_test = train_test_split(
    X,
    y,
    test_size=0.2,
    stratify=y,
    random_state=42
)
```

```python
#confirming stratified train test split worked
y_train.value_counts(normalize=True)*100
y_test.value_counts(normalize=True)*100
```

Above test performed show the that the split was successful

```python
from sklearn.pipeline import Pipeline
from sklearn.preprocessing import StandardScaler
from sklearn.linear_model import LogisticRegression

lr_pipeline = Pipeline([
    ("scaler", StandardScaler()),
    ("model", LogisticRegression(
        class_weight="balanced",
        max_iter=1000,
        random_state=42
    ))
])
```

Pipelines are used to ensure that there is no data leakage and to make the code reusable
